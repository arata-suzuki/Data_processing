<h1><center> Est-il possible d'investir dans le BTC au bon moment en suivant le modele de certains Acteurs? </center></h1>

Le "mirror trading" est une stratégie d'investissement qui consiste à imiter les actions de traders expérimentés ou d'autres investisseurs performants en reproduisant leurs transactions sur les marchés financiers. Durant la période de 2015 à 2017, certains investisseurs ont su faire preuve de timing et de discernement dans leurs décisions d'achat et de vente, ce qui a entraîné des gains significatifs. En identifiant ces acteurs et en analysant leur comportement d'investissement, il est possible de tirer des enseignements pour optimiser ses propres investissements.

La pratique du "mirror trading" repose sur l'idée que certains investisseurs ont une compréhension approfondie des marchés financiers, des tendances économiques et des facteurs qui influencent les prix des actifs. En suivant leurs transactions, on espère bénéficier de leur expertise et de leur capacité à prendre des décisions éclairées. Les investisseurs qui réussissent à maximiser leurs achats en achetant des actifs sous-évalués ou sur le point de connaître une hausse de valeur, et en vendant des actifs surévalués ou sur le point de connaître une baisse de valeur, sont particulièrement intéressants à imiter.

Pour mettre en pratique le "mirror trading", il est nécessaire d'identifier les investisseurs performants, de suivre leurs transactions en temps réel et de reproduire leurs actions sur les marchés financiers. Il est également important de s'assurer que les investisseurs que l'on suit sont en mesure de prendre des décisions éclairées et de ne pas se laisser influencer par des facteurs émotionnels. Enfin, il est nécessaire de s'assurer que les investisseurs que l'on suit sont capables de prendre des décisions indépendamment les uns des autres, afin d'éviter de reproduire les mêmes transactions que d'autres investisseurs. 


En somme, le "mirror trading" est une stratégie d'investissement qui consiste à imiter les actions d'investisseurs performants dans le but d'obtenir un gain assuré en reproduisant leur comportement d'investissement.

Nous allons donc essayer d'identifier de tels acteurs.

On définit un moment clé d'investissement selon 2 critéres :  
    - le BTC a atteint un minimu local  
    - passer le miminum le BTC n'a fait que augmenter sur 7 jours.

Avant de commencer, nous allons charger les données nécessaires, notamment celles de "blockchain_by_actor" qui nous permettront de savoir qui a acheté et vendu des bitcoins, ainsi que celles de "external" qui décrivent le cours du bitcoin sur la période de 2015 à 2017.

Le jeu de données "blockchain_by_actor" contient des informations sur les transactions de bitcoins effectuées par différents acteurs sur la blockchain. Ces données nous permettront de déterminer qui a acheté et vendu des bitcoins au cours de la période d'étude.

Le deuxième jeu de données, "external", fournit des données sur le cours du bitcoin au cours de la période de 2015 à 2017. Ces données nous permettront de suivre les variations du prix du bitcoin pendant cette période et de les corréler avec les transactions effectuées par les acteurs de la blockchain.

Une fois que nous aurons chargé ces données, nous serons en mesure de procéder à une analyse approfondie pour comprendre les tendances d'achat et de vente de bitcoins, ainsi que les relations entre ces transactions et les fluctuations du prix du bitcoin sur le marché externe. Cela nous permettra de tirer des conclusions pertinentes sur les activités de la blockchain et les comportements des acteurs impliqués dans les transactions de bitcoins.

```python
#Global Values

path = "DefiEGC2024-main\\timeseries\\"
excelActor = pd.read_csv(path + "blockchain_by_actor.csv")
excelExternal = pd.read_csv(path + "external.csv")


unitSatochi = 0.00000001
INTERESTPOINTLIMIT = 10

valExternal = excelExternal.PriceUSD.to_list()
datesExternal = excelExternal.date.to_list()
actorsName = np.unique(excelActor.identity.values) #nom des acteurs
```

Après avoir chargé les données, nous analysons le cours du bitcoin sur la période de 2015 à 2017 pour repérer les moments clés d'investissement. Nous identifions les moments où le bitcoin atteint un minimum local et où il augmente sur une période de 7 jours, et nous stockons ces moments dans un tableau. De même, nous repérons les moments où le bitcoin atteint un maximum local et où il diminue sur une période de 7 jours, et nous stockons ces moments dans un autre tableau. Ainsi, nous obtenons les moments propices à l'achat et ceux où il est préférable de ne pas investir.

```python
#BTC display

NBRvalExternal = len(valExternal)

#fonction qui permet de recuperer les moments où la valeur du bitcoin croit sur une periode minimum de 7 jours
def isUseFull(listIndice):
    acceptable = 7
    list = listIndice.tolist()
    cpylist = [x for x in list]
    for ind in cpylist:
        if (ind + acceptable < NBRvalExternal - 1):
            for k in range(1,acceptable):
                diff = valExternal[ind + k - 1] - valExternal[ind + k ] 
                if (diff > 0):
                    list.remove(ind)
                    break
    print("nouveau minimum locaux : " , len(list))
    return list

#fonction qui permet de recuperer les moments où la valeur du bitcoin decroit sur une periode minimum de 6 jours
def isworst(listIndice):
    acceptable = 6
    list = listIndice.tolist()
    cpylist = [x for x in list]
    for ind in cpylist:
        if (ind + acceptable < NBRvalExternal - 1):
            for k in range(1,acceptable):
                diff = valExternal[ind + k - 1] - valExternal[ind + k ] 
                if (diff < 0):
                    list.remove(ind)
                    break
    print("nouveau maximum locaux : " , len(list))
    return list

#fonction qui permet de recuperer les moments les plus interressant de l'evolution de la valeur du BTC, c'est à dire les moments avant lesquels investir
def MostInterstedMoment():
    min_indices = argrelextrema(np.array(valExternal), np.less)
    print("minimum locaux : " , len(min_indices[0])) #bcp trop de min local recuperer les plus interressant se limiter a une dizaine 
    min_indices = isUseFull(min_indices[0])
    return [datesExternal[ind] for ind in min_indices]

def MostInterstedMomentIndice():
    min_indices = argrelextrema(np.array(valExternal), np.less)
    return isUseFull(min_indices[0])

#fonction qui permet de recuperer les moments les moins interressant de l'evolution de la valeur du BTC, c'est à dire les moments avant lesquels il ne faut pas investir
def worstInterstedMoment():
    max_indices = argrelextrema(np.array(valExternal), np.greater)
    print("maximum locaux : " , len(max_indices[0])) #bcp trop de max local recuperer les plus interressant se limiter a une dizaine 
    max_indices = isworst(max_indices[0])
    max_indices.append(530)
    max_indices.sort()
    return [datesExternal[ind] for ind in max_indices][:-1]

#fonction qui permet d'afficher les meilleurs et les pires moments pour investir sur la courbe de l'evolution de la valeur du BTC
def BTCCourbes():
    fig, ax = plt.subplots(figsize = (30, 12))
    
    plt.plot(datesExternal, valExternal, color='orange' ,  label='BTC')
    plt.title("BTC evolution")

    mindatesExternal = MostInterstedMoment()
    maxdatesExternal = worstInterstedMoment()
    fmi,fma = True, True
    for date in mindatesExternal:
        if fmi:
            plt.plot([date for x in range(int(max(valExternal)))], [x for x in range(int(max(valExternal)))], '--',color='green', label = 'Most intersting Moment')
            fmi = False
        else:
            plt.plot([date for x in range(int(max(valExternal)))], [x for x in range(int(max(valExternal)))], '--',color='green')
    for date in maxdatesExternal:
        if fma: 
            fma = False
            plt.plot([date for x in range(int(max(valExternal)))], [x for x in range(int(max(valExternal)))], '--',color='red', label = 'worst intersting Moment')
        else:
            plt.plot([date for x in range(int(max(valExternal)))], [x for x in range(int(max(valExternal)))], '--',color='red')
    plt.xticks(np.arange(0, 900, 40))
    ax.xaxis.set_tick_params(rotation=45, labelsize=10)
    plt.legend(prop={'size': 15})
    plt.show()


#fonction qui permet d'afficher les 10 meilleurs investisseurs
def displayMostActors(listTuple):
    fig, ax = plt.subplots(nrows=2, ncols=4, figsize = (25, 12))
    col = 0
    for i in range(len(listTuple)):
        date , dataFrame = listTuple[i]
        Actors = dataFrame.identity.to_list()[:10]
        recieved = dataFrame.received.to_list()[:10]
        ax[col, i%4].bar(Actors, recieved)
        ax[col, i%4].xaxis.set_tick_params(rotation=45, labelsize=10)
        ax[col, i%4].set_ylabel('BTC Received')
        ax[col, i%4].set_title('Most important investisor : ' + date )
        if (i == 3):
            col += 1
    plt.tight_layout()
    plt.show()

#fonction qui permet d'afficher les 10 pires investisseurs
def displayWorstActors(listTuple):
    fig, ax = plt.subplots(nrows=2, ncols=4, figsize = (25, 12))
    col = 0
    for i in range(len(listTuple)):
        date , dataFrame = listTuple[i]
        Actors = dataFrame.identity.to_list()[:10]
        recieved = dataFrame.received.to_list()[:10]
        ax[col, i%4].bar(Actors, recieved)
        ax[col, i%4].xaxis.set_tick_params(rotation=45, labelsize=10)
        ax[col, i%4].set_ylabel('BTC Received')
        ax[col, i%4].set_title('Worst important investisor : ' + date )
        if (i == 3):
            col += 1
    plt.tight_layout()
    plt.show()

BTCCourbes()
```

Nous identifions 8 moments clés d'investissement en analysant les données du cours du bitcoin (BTC) sur la période de 2015 à 2017. Maintenant, nous devons déterminer qui a investi à ces moments.

```python
#fonction qui permet de recuperer les meilleurs investisseurs, et calculer le nombre de BTC qu'ils ont recu
def Mostreceived(importantDates):
    res = []
    for date in importantDates:
        ReceivedBydate = excelActor.loc[excelActor['date'] == date][['identity', 'received']].sort_values('received',ascending=False)
        ReceivedBydate.received = ReceivedBydate.received * unitSatochi
        res.append((date, ReceivedBydate))
    displayMostActors(res)

Mostreceived(MostInterstedMoment())
```

```python
#fonction qui permet de monter la fréquence d'achat du BTC par les acteurs récupéré ci-dessus
def displayHistogram(data, title):
    bins = 50
    plt.figure(figsize=(30,15))    
    plt.hist(data[0], bins, alpha=0.5, label=title[0])
    plt.hist(data[1], bins, alpha=0.5, label=title[1])
    plt.hist(data[2], bins, alpha=0.5, label=title[2])

    plt.xlabel('Actors')
    plt.ylabel('Fréquence')
    plt.title('Classement des Acteurs ayant le plus acheter de BTC')
    plt.xticks(rotation = 35)
    plt.legend(loc='upper right', title_fontsize = 5, prop={'size': 20})
    plt.show()

#fonction qui permet de recuperer le nombre de fois où les acteurs sont arrivés en 1er, 2eme et 3eme position en terme de recette
def mostReceived():
    res1,res2,res3 = [],[],[]
    for date in datesExternal:
        ReceivedBydate = excelActor.loc[excelActor['date'] == date][['identity', 'received']].sort_values('received',ascending=False)
        ReceivedBydate.received = ReceivedBydate.received * unitSatochi
        res3.append(ReceivedBydate.values[2][0])
        res2.append(ReceivedBydate.values[1][0])
        res1.append(ReceivedBydate.values[0][0])


    return res1,res2,res3

x1,x2,x3 = mostReceived()

displayHistogram([x1,x2,x3], ['First','Second','Third'])
```

```python
StatsByActors = []
for name in actorsName :
    mean = round(excelActor[excelActor.identity == name]['received'].mean() * unitSatochi, 1)
    ecartType = round(excelActor[excelActor.identity == name]['received'].std() * unitSatochi, 1)
    StatsByActors.append((name,  mean, ecartType))


def tri_tupes(lst):
    return lst[2] - lst[1]

#biggest ecart type
StatsByActorssortedBySTD =  sorted(StatsByActors, key=tri_tupes, reverse=True)
print("classement par ecart type maximiser par rapport a la moyenne")
for i in range(10):
    print(i + 1, " : ", StatsByActorssortedBySTD[i][0], "   ",StatsByActorssortedBySTD[i][2])

```
