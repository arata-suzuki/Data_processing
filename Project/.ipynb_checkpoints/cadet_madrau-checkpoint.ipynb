## <u>Problématique :</u> Analyse de l'impact de la puissance de hachage sur les frais de transaction dans la blockchain Bitcoin

Nous voulons étudier la relation entre la puissance de hachage et les frais de transaction dans la blockchain Bitcoin.

### Le HashRate

La puissance de hachage, mesurée en `HashRate` dans le fichier `external.csv`, correspond à la puissance de calcul totale utilisée par les mineurs sur le réseau Bitcoin pour résoudre les opérations (hashs) nécessaires à la validation des transactions et à la création de nouveaux blocs de transaction.

Plus la puissance de hachage est élevée, plus le réseau est sécurisé et résistant aux attaques.

### Les frais de transactions

Les frais de transaction, mesurés en  `mean_feeUSD` dans le fichier `global.csv`, correspondent aux montants payés par les utilisateurs pour que leurs transactions soient traitées rapidement par les mineurs et ajoutées à la blockchain. Les mineurs traitent les transactions avec les frais les plus élevés en priorité, car cela leur permet de gagner plus grâce aux frais de transaction.

### Y-a-t'il une relation entre les deux ?

En étudiant la relation entre la puissance de hachage et les frais de transaction, **il serait possible de déterminer si une augmentation de la puissance de hachage entraîne une augmentation des frais de transaction dans la blockchain Bitcoin**, en raison de la concurrence accrue entre les utilisateurs pour que leurs transactions soient traitées rapidement.

## 3. Nettoyer le DataFrame

Nous remarquons que la courbe concernant les frais de transaction contient des pics. Ces pics sont des erreurs, aucun article sur internet ne justifie leur existence.
Nous cherchons donc à les identifier pour les supprimer.

Nous voulons, par exemple, identifier la date des deux gros pics de la première moitié de la courbe.

```python
# Search of the date of the two bigs peaks before 2016-06-01
df[df.index < '2016-06-01'].sort_values(by='mean_fee_satoshi', ascending=False).head(2)
```

Ces pics ont eu lieu les 25 avril 2015 et 26 avril 2016.

Une façon d'identifier ces pics brusques est de calculer le `diff` sur la colonne `mean_fee_satoshi`.

Le `diff` correspond à la différence entre une valeur et la valeur précédente. Ainsi, lorsqu'il y a un pic, la différence entre les deux valeurs est très importante.

Nous cherchons donc à identifier les valeurs pour lesquelles le `diff` est très important afin d'identifier les erreurs de notre DataFrame actuel.

```python
# We calculate the `diff` of values between 2016-04-22 and 2016-04-30
pd.DataFrame(df[(df.index < '2016-04-30') & (df.index > '2016-04-22')]['mean_fee_satoshi'].diff())
```

Nous remarquons que le `diff` du 26 avril est très grand : 114 466 satoshis.

Nous savons donc que nous devons enlever cette valeur.

Ainsi, il suffit de supprimer les lignes dont le `diff` est supérieur à un certain seuil, que nous prenons à 10 000.

```python
# We delete values with a `diff` > 10 000
df.drop(df.index[df["mean_fee_satoshi"].diff() > 10_000], inplace = True)
```

```python
fig, ax1 = plt.subplots()
ax1.tick_params(axis='x', rotation=45)

# HashRate plot
ax1.set_ylabel('hash/s', color='tab:red')
ax1.plot(df["HashRate"], color='tab:red', label='HashRate')
ax1.legend(loc=0)
ax1.tick_params(axis='y', labelcolor='tab:red')


ax2 = ax1.twinx()

# Mean fee plot
ax2.set_ylabel('satoshis', color='tab:blue')
ax2.plot(df["mean_fee_satoshi"], color='tab:blue', label="Frais")
ax2.legend(loc=1)
ax2.tick_params(axis='y', labelcolor='tab:blue')

fig.tight_layout()
plt.title("Évolution de la puissance de hachage (rouge) et frais de transaction (bleu) en fonction du temps")
plt.show()
```

Notre courbe des frais de transaction est maintenant plus exploitable sans ces pics qui pouvaient causer des erreurs dans notre étude.

## 4. Analyse brute des données

Dans cette partie, nous analysons les données de notre DataFrame `df`.

```python
plt.title("Frais de transaction en fonction de la puissance de hachage")
plt.plot(df["HashRate"], df["mean_fee_satoshi"], '.')
plt.xlabel("HashRate"); plt.ylabel('Frais de transaction')
plt.show()
```

Lorsque nous observons la courbe des frais de transactions en fonction de la puissance de hachage, nous remarquons que celle-ci a tendance à croître.

Cependant, si la tendance observée est monotone, elle n'est pas affine. Nous allons donc utiliser [le coefficient de Spearman](https://en.wikipedia.org/wiki/Spearman%27s_rank_correlation_coefficient) et non [le coefficient de Pearson](https://en.wikipedia.org/wiki/Pearson_correlation_coefficient) pour étudier leur corrélation. 

Le coefficient de corrélation est une valeur comprise entre -1 et 1. Plus le coefficient de corrrélation est proche de 1, plus les données sont statistiquement corrélées.

```python
# Study of the correlation of our datas with Spearman's coefficient
df.corr(method = "spearman").style.background_gradient(cmap='coolwarm')
```

Entre la puissance de hachage et les frais de transaction, nous obtenons un coefficient de corrélation de 0.91 .
Vu que le coefficient de corrélation est très proche de 1, on peut donc dire que les données sont fortement corrélées.

Ce résultat appuie notre hypothèse que **la puissance de hachage et les frais de transaction sont liés.**

Cependant, lorsque nous obervons les courbes, nous remarquons qu'elles présentent des perturbations qui sont suceptibles de fausser le calcul de la corrélation. Afin de renforcer notre hypothèse, nous allons tenter dans la prochaine partie de réduire ces perturbations en lissant les valeurs.

___

## 5. Analyse des données lissées

Dans cette partie, nous analysons les données de notre DataFrame `df` que nous avons lissées. Lisser des valeurs permet de réduire les perturbations et le bruit sur une série de valeurs afin de pouvoir observer une tendance générale.

Pour ce faire, nous remplaçons chaque valeur par sa moyenne glissante, c'est à dire la moyenne des $n$ valeurs précédantes et suivantes.

```python
def smooth_dataframe(df, dict_n):
    # Create a new DataFrame to hold the smoothed values
    smoothed_df = pd.DataFrame(index=df.index, columns=df.columns)

    # Loop over each column in the DataFrame
    for col in df.columns:
        n = dict_n[col]
        # Smooth the column using a rolling window of size 2*n+1
        smoothed_col = df[col].rolling(2*n+1, center=True).mean()

        # Fill in any NaN values at the beginning and end of the smoothed column
        smoothed_col[:n] = np.nan
        smoothed_col[-n:] = np.nan

        # Copy the smoothed column into the new DataFrame
        smoothed_df[col] = smoothed_col

    return smoothed_df
```

- Pour la puissance de hachage, nous remplaçons par la moyennes des 3 valeurs autour.
- Pour les frais de transaction, nous remplaçons par la moyenne des 7 valeurs autour.

```python
# New DataFrame
df_smooth = smooth_dataframe(df, {'HashRate': 3, 'mean_fee_satoshi': 7})
```

```python
fig, ax1 = plt.subplots()
ax1.tick_params(axis='x', rotation=45)

# HashRate plot
ax1.set_ylabel('hash/s', color='tab:red')
ax1.plot(df_smooth["HashRate"], color='tab:red', label='HashRate')
ax1.legend(loc=0)
ax1.tick_params(axis='y', labelcolor='tab:red')

ax2 = ax1.twinx()

# Mean fee plot
ax2.set_ylabel('satoshis', color='tab:blue')
ax2.plot(df_smooth["mean_fee_satoshi"], color='tab:blue', label='Frais')
ax2.legend(loc=1)
ax2.tick_params(axis='y', labelcolor='tab:blue')

fig.tight_layout()
plt.title("Évolution de la puissance de hachage (rouge) et frais de transaction (bleu) en fonction du temps\nValeurs lissées")
plt.show()
```

Nous remarquons que les courbes présentent nettement moins de perturbations.

Réaffichons les frais de transaction en fonction de la puissance de hachage comme précédemment, mais cette fois-ci avec les valeurs lisées:

```python
plt.title("Frais de transaction en fonction de la puissance de hachage\nValeurs lissées")
plt.plot(df_smooth["HashRate"], df_smooth["mean_fee_satoshi"], '.')
plt.xlabel("HashRate"); plt.ylabel('Frais de transaction')
plt.show()
```

La courbe présente quelques défauts, dûs aux imprécisions de la moyenne, mais nous observons bien une tendance générale croissante, comme supposée précédemment.

Maintenant que nous avons réduit le bruit de nos valeurs, nous pouvons recalculer la corrélation: 

```python
# Study of the correlation of our datas with Spearman's coefficient
df_smooth.corr(method = "spearman").style.background_gradient(cmap='coolwarm')
```

Avec les données lissées, le coefficient de corrélation augmente pour arriver à 0.94. Il s'est donc rapproché de 1, ce qui renforce notre hypothèse que la puissance de hachage influe sur les frais de transaction.

## 6. Conclusion de l'étude

Après avoir étudié une possible relation entre la puissance de hachage et les frais de transaction, nous pouvons déterminer si une augmentation de la puissance de hachage entraîne une augmentation des frais de transaction dans la blockchain Bitcoin.

**Nous avons pu montrer que la puissance de hachage est fortement corrélée avec les frais de transaction :**
 - Le coefficient de corrélation est de **0.91** avec les données brutes ;
 - Le coefficient de corrélation est de **0.94** avec les données lissées.

**Nous en déduisons qu'il y a bien une relation entre la puissance de hachage et les frais de transaction.**

Cependant, nous savons que la corrélation ne prouve pas la causalité : d'autres facteurs peuvent également contribuer à l'augmentation des frais de transaction. 

Néanmoins, notre étude suggère qu'il y a bien une relation entre la puissance de hachage et les frais de transaction, et que l'augmentation de la puissance de hachage peut potentiellement augmenter la concurrence entre les utilisateurs pour que leurs transactions soient traitées rapidement, ce qui peut à son tour entraîner une augmentation des frais de transaction.
